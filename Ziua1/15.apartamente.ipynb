```python
from scipy import stats
```

```python
stats_results = stats.linregress(suprafete, preturi)
stats_results.slope, stats_results.intercept
```

```python


plt.scatter(suprafete, preturi)
plt.plot(suprafete, preturi)

rezultat_prezis1 = 1000 * suprafete + 1000
plt.plot(suprafete, rezultat_prezis1)

rezultat_prezis2 = 1300 * suprafete + 1000
plt.plot(suprafete, rezultat_prezis2)

rezultat_optim = 1225.6249497628808 * suprafete + 2895.940840768424
plt.plot(suprafete, rezultat_optim)
```

```python
plt.scatter(suprafete, preturi)
plt.plot(suprafete, preturi)


rezultat_optim = stats_results.slope * suprafete + stats_results.intercept
plt.plot(suprafete, rezultat_optim)
```
